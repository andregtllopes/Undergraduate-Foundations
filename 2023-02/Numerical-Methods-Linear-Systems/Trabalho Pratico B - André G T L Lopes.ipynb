```python
def entrada():
    """
    Solicita ao usuário a quantidade de equações do sistema,
    obtém os coeficientes e termos independentes e verifica
    a validade das entradas.

    Returns:
    A tuple (A, B) com as listas de coeficientes e termos independentes.
    """
    true = True # Uso de variável booleana para evitar repetições infinitas.
    while true: # Loop de verificação de entrada
        try:
            # Solicita ao usuário a quantidade de equações e verifica se é um inteiro positivo.
            entrada_usuario = input("Forneça a quantidade de equações do sistema: ") # Solicita ao usuário inserir a quantidade de equações do sistema e armazena a entrada na variável 

            if "." in entrada_usuario: #Verifica se há um ponto decimal na entrada do usuário.
                print("Por favor, insira um número inteiro positivo.") # Se o ponto decimal estiver presente, imprime uma mensagem solicitando um número inteiro positivo.
                continue  # Volta para o início do loop se a entrada contiver um ponto decimal (indicando um float).

            numero_de_equacoes = int(entrada_usuario) #transforma o valor para int, para que possa ser utilizado.

            if numero_de_equacoes <= 0: #número de equações não pode ser igual a zero ou negativo.
                print("O número de equações deve ser um inteiro maior que zero.") # Se o número de equações for inválido, imprime uma mensagem de erro.
            else: # Se o número de equações for válido (inteiro positivo), executa o bloco de código a seguir.
                true = False  #  Altera o valor de true para False, saindo do loop while.
        except ValueError: #Captura uma exceção do tipo ValueError que pode ocorrer se a conversão para inteiro falhar.
            print("Por favor, insira um número inteiro positivo.") # Imprime uma mensagem de erro para o caso em que a exceção ValueError ocorre.

    listacoeficientes = [] # Inicializa a lista que conterá os coeficientes do sistema.

    for i in range(1, numero_de_equacoes + 1):# Loop para obter os coeficientes de cada equação.
        linha_coef = [] #lista temporária que armazena os coeficientes de uma única linha da matriz.
        print(f"\nInsira os coeficientes da equação {i} (total de {i} coeficientes esperados):") #Solicita ao usuário inserir os coeficientes da equação atual e exibe uma mensagem informativa.
        for j in range(1, i + 1): # Inicia um loop para obter os coeficientes de cada coluna da equação atual. O loop percorre as colunas de 1 até i.
            while True: # Inicia um loop infinito para garantir que o usuário forneça um coeficiente válido.
                try: # Inicia um bloco de código que pode gerar exceções.
                    coeficiente = float(input(f"Coeficiente {j} da equação {i}: ")) # Solicita ao usuário o coeficiente da linha i, coluna j.
                except ValueError: #Captura a exceção ValueError que ocorre se a conversão falhar.
                    print("O coeficiente deve ser um número real. Tente novamente.") #Imprime uma mensagem de erro para o usuário.
                    continue # Volta para o início do loop while, pedindo novamente ao usuário que insira o coeficiente.
                else: # Se nenhum erro ocorreu durante o bloco try, executa o bloco de código a seguir.
                    break # Sai do loop while e continua para a próxima iteração do loop for.
            linha_coef.append(coeficiente) # Adiciona o coeficiente à lista temporária linha_coef.
        listacoeficientes.extend(linha_coef) # Estende a lista listacoeficientes com os coeficientes da linha atual.

    # O código verifica se a matriz é triangular inferior, e se não for, imprime uma mensagem de erro e reinicia o processo.
    for i in range(numero_de_equacoes):
        if listacoeficientes[int(i * (i + 1) / 2) + i] == 0:
            print("Os coeficientes de valor de linha e coluna iguais não podem ser zero. Reinicie o processo.")
            return entrada()

    listatermosindependentes = [] #Inicializa a lista que conterá os termos independentes.

    for i in range(numero_de_equacoes): #Inicia um loop para obter os termos independentes de cada equação.
        while True: # Inicia um loop infinito para garantir que o usuário forneça um termo independente válido.
            try:
                # Solicita ao usuário o termo independente da equação i.
                termo_independente = float(input(f"Termo independente da equação {i + 1}: ")) #Solicita ao usuário inserir um termo independente e converte a entrada para um número de ponto flutuante.
            except ValueError: # Captura a exceção ValueError que ocorre se a conversão falhar.
                print("O termo independente deve ser um número real. Tente novamente.") # Imprime uma mensagem de erro para o usuário.
                continue #Volta para o início do loop while, pedindo novamente ao usuário que insira o termo independente.
            else: #Se nenhum erro ocorreu durante o bloco try, executa o bloco de código a seguir.
                break #Sai do loop while e continua para a próxima iteração do loop for.
        listatermosindependentes.append(termo_independente) # Adiciona o termo independente à lista listatermosindependentes.

    print(f"\nCoeficientes: {listacoeficientes}\nTermos Independentes: {listatermosindependentes}\n") #Imprime as listas de coeficientes e termos independentes após a conclusão do processo.

    confirmação_para_passar = True #Inicializa uma variável chamada confirmação_para_passar com o valor True.
    while confirmação_para_passar: #Inicia um novo loop while que continuará até que confirmação_para_passar seja alterada para False.
        confirmar = int(input("Confirme os valores digitando 1. Para retornar, digite 0:")) #Solicita ao usuário que confirme os valores digitando 1 ou retorne digitando 0.
        if confirmar in {0, 1}: #Verifica se a entrada do usuário é 0 ou 1.
            confirmação_para_passar = False #Altera o valor de confirmação_para_passar para False, saindo do loop while.
    if confirmar == 1: #Se a confirmação do usuário for 1, executa o bloco de código a seguir.  
        A = listacoeficientes #Atribui a lista listacoeficientes à variável A.
        B = listatermosindependentes # Atribui a lista listatermosindependentes à variável B.

        return A, B # Retorna as listas A e B.
    else: # Se a confirmação do usuário for 0, executa:
        return entrada() # Chama recursivamente a função entrada() para reiniciar o processo.

def sistema_triangular(A, B): # Define a função sistema_triangular que recebe duas listas como parâmetros, A e B.
    """
    Resolve um sistema de equações lineares representado por uma matriz
    triangular inferior.

    Args:
    A (list): Lista de coeficientes do sistema (matriz triangular inferior).
    B (list): Lista de termos independentes.

    Returns:
    Uma lista X com as soluções do sistema.
    """
    n = len(B)  # Número de equações (e, portanto, de termos independentes)
    X = [0.0] * n #Inicializa a lista X que conterá as soluções, preenchida com zeros.

    # Construção da matriz diretamente
    Matriz = []  # Inicializa a lista que conterá a matriz triangular inferior
    for i in range(n): # Loop externo percorrendo cada linha da matriz.
        linha = [0.0] * n  # Inicializa cada linha com n zeros
        for j in range(i + 1): # Loop interno percorrendo cada coluna da matriz.
            linha[j] = A[int(i * (i + 1) / 2) + j] # Preenche a linha com os coeficientes da lista A seguindo a lógica da matriz triangular inferior.
        Matriz.append(linha) # Adiciona a linha preenchida à matriz.

    # Resolução do sistema triangular
    for i in range(n): # Loop externo para resolver o sistema linear usando a matriz triangular inferior.
        somatorio = 0 # Inicializa a variável somatorio que será usada para calcular a soma dos produtos dos coeficientes.
        for j in range(i): # Loop para calcular o somatório dos produtos dos coeficientes da matriz pelo vetor de soluções já encontradas.
            somatorio += Matriz[i][j] * X[j] #Calcula e acumula os produtos dos coeficientes da matriz pelo vetor de soluções.
        X[i] = (B[i] - somatorio) / Matriz[i][i] # Calcula o valor da incógnita atual usando o termo independente e o somatório, dividido pelo coeficiente associado à incógnita na matriz triangular.

    # Formatação dos resultados para até 3 casas decimais
    X = ["%.3f" % valor for valor in X]

    return X # Função retorna uma lista X com os valores calculados para as incógnitas.
(A, B) = entrada() #Nesse momento é feita a associação entre as funções, já que 
X = sistema_triangular(A,B)
def exibe(A, B, X): #Define a função exibe que recebe três listas como parâmetros, A, B e X.
    """
    Exibe o sistema de equações e as soluções.

    Args:
    A (list): Lista de coeficientes do sistema (matriz triangular inferior).
    B (list): Lista de termos independentes.
    X (list): Lista de soluções do sistema.

    Returns:
    Uma string com a representação do sistema e suas soluções.
    """
    n = len(B) #Obtém o número de equações (ou termos independentes) do sistema.

    Matriz = [] #Inicializa uma lista vazia que será preenchida com as linhas da matriz triangular inferior.
    for i in range(n): #Inicia um loop externo para percorrer cada linha da matriz triangular inferior.
        linha = [] # Inicializa uma lista vazia que representará uma linha da matriz.
        for j in range(n): # Inicia um loop interno para percorrer cada coluna da matriz.
            if j <= i: # Verifica se a coluna é menor ou igual à linha para preencher a matriz triangular inferior.
                linha.append(float(A[int(i * (i + 1) / 2) + j])) #Adiciona o coeficiente correspondente à posição (i, j) da matriz triangular inferior.
            else: #Caso contrário, preenche com 0.0.
                linha.append(0.0)
        Matriz.append(linha) # Adiciona a linha preenchida à matriz.

    exibicao_sistema = "" #Inicializa uma string vazia para armazenar a exibição do sistema de equações.

# Exibindo o sistema de equações
    for i in range(n): #Inicia um loop externo para exibir o sistema de equações.
        linha = "" # Inicializa uma string vazia para representar uma linha do sistema.
        for j in range(n): # Inicia um loop interno para percorrer cada coluna da matriz.
            if Matriz[i][j] != 0: #Verifica se o coeficiente é diferente de zero para incluir no sistema.
                if linha != "": # Adiciona um sinal de + se a linha não estiver vazia.
                    linha += " + "
                linha += f"({Matriz[i][j]:.3f})*X{j + 1}" # Cria o termo formatado para ser adicionado à linha.
            else:
                linha += " + 0.0" # Se o coeficiente for zero, adiciona " + 0.0".
        exibicao_sistema += linha + f" = {B[i]:.3f}\n" # Adiciona a linha do sistema com o termo independente à exibição do sistema.

    # Exibindo as soluções
    solucoes = "" # Inicializa uma string vazia para representar as soluções.
    for j in range(n): # Inicia um loop para exibir as soluções.
        solucoes += f"X{j + 1} = {float(X[j]):.3f}" #Adiciona a solução formatada à string.
        if j < n - 1: #Adiciona espaços entre as soluções, exceto para a última.
            solucoes += "  "

    exibicao_sistema += solucoes #Adiciona as soluções à exibição do sistema.

    print("\nO sistema está pronto e resolvido!\n") #Exibe uma mensagem indicando que o sistema está pronto e resolvido.

    return exibicao_sistema #Retorna a exibição do sistema.

print(exibe(A, B, X))



```
